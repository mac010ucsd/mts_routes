```python
import networkx as nx
import osmnx as ox

ox.__version__
```

```python
# G = ox.graph.graph_from_place("San Diego, California, USA", network_type="walk")
G = ox.graph.graph_from_point(ox.geocoder.geocode("UCSD"), 1000, network_type="walk")
```

```python
fig, ax = ox.plot.plot_graph(G, figsize=(10,10))
```

```python
G1 = ox.routing.add_edge_speeds(G)
G1 = ox.routing.add_edge_travel_times(G1)
```

```python
# nx.get_edge_attributes(G1)

#G1.edges(data=True)
```

```python
ec = ox.plot.get_edge_colors_by_attr(G, "travel_time", cmap="inferno")
fig, ax = ox.plot.plot_graph(G1, edge_color=ec, edge_linewidth=2, node_size=0)
```

```python
G2 = ox.convert.to_undirected(G1)
```

```python

data_flow = nx.current_flow_closeness_centrality(G2, weight="travel_time")
```

```python
# nx.Graph(data_flow)
```

```python
nx.set_node_attributes(G2, data_flow, name="flow")
# G3
```

```python
nc = ox.plot.get_node_colors_by_attr(G2, "flow", cmap="inferno")
fig, ax = ox.plot.plot_graph(G2, node_color=nc)
```
